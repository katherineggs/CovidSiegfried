```python
import pandas as pd
import datetime
```

# Confirmed

```python
# --- read data
dfConfirmed = pd.read_csv('monitor/time_series_covid19_confirmed_global.csv')
dfConfirmed
```

```python
# --- order data
conf = dfConfirmed.melt(id_vars=["Province/State", "Country/Region", "Lat", "Long"], 
                        var_name="Date", 
                        value_name="Confirmed")

#to date type
conf['Date'] = pd.to_datetime(conf['Date'])

#add new columns
conf['Year'] = conf['Date'].dt.year 
conf['Month'] = conf['Date'].dt.month
conf['Day'] = conf['Date'].dt.day

conf
```

```python
#Rename columns
conf = conf.rename(columns={'Long': 'Lon'})
conf
```
